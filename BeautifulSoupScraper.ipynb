```python
names = []
emails = []
```

```python
for name in member_info:
    url = "https://estore.reit.com/Membership/IndividualInformation.aspx?" + member_info[name]
    #print(url)
    member_page = requests.get(url)
    soup = BeautifulSoup(member_page.text, 'html.parser')
    information = soup.find("td", {"class":"isg_formLabel"})
    person = information.find("b").text.strip()
    print(person)
    email = information.find("a").text
    names.append(person)
    emails.append(email)
```
